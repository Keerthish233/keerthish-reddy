#### Assumption in Multilinear Regression

1. Linearity: The relationship between the predictors and the response is linear.

2.Independence: Observation are independent of each other.

3.Homoscedasticity: The residuals (Y - Y hat)) exhibit constant variance at all levels of the predictor.

4.Normal Distribution of Errors: The residuals of the model are normally distributed.

5.No multicollinearity: The independent variables should not be too highly correlated with each other

#### Preparing a preliminary model considering all x columns

```python
#Bulid model
#import statsmodels.formula.api as smf
model = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model.summary()
```

```python
#Bulid model
import statsmodels.formula.api as smf
model1 = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
```

```python
model1.summary()
```

#### Observations from model summary:
1. The R-squared and adjusted R-squared values are good and about 75% of variability in y is explianed by x columns
2. The probabaility value with respect to F_stastic is close to zero indicating that all are some of x columns are significant
3. The p-values for VOL and WT are higer than 5% indicating some intercation issue among themselves which need to be further explored

#### Performance metrics for model1

```python
#Find the performance metrics
#Create a data frame with actual y and predicted y columns

df1 = pd.DataFrame()
df1["actual_y1"] = cars["MPG"]
df1.head()
```

```python
# Predict for the given x data columns
pred_y1 = model1.predict(cars.iloc[:,0:4])
df1["pred_y1"] = pred_y1
df1.head()
```

```python
# Compute the ,mean Squared Error (MSE) for model1

from sklearn.metrics import mean_squared_error
print("MSE :", mean_squared_error(df1["actual_y1"], df1["pred_y1"]))
```
